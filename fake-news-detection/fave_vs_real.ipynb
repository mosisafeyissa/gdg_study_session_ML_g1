```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_excel("fake_vs_real_news.xlsx")
df.head()
```

```python
df['label'].value_counts()
```

```python
X = df['text']
y = df['label']

vectorizer = TfidfVectorizer(max_features=5000)
X_vectorized = vectorizer.fit_transform(X).toarray()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized, y,
    test_size=0.2,
    random_state=42
)
```

```python
model = Sequential()

model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=15,
    batch_size=32,
    validation_data=(X_test, y_test)
)
```

## Model Accuracy result

```python
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", accuracy)
```

```python
def predict_news(text):
    text_vec = vectorizer.transform([text]).toarray()
    prediction = model.predict(text_vec)

    if prediction[0][0] > 0.5:
        print("Fake News")
    else:
        print("Real News")
```

```python
predict_news("Scientists confirm water on Mars")
predict_news("Secret government project creates invisible humans")
```

```python
model.save("fake_news_model.keras")
```

## Neural Network Architecture
I used a Feedforward Neural Network for text classification and the model consists:
An input layer that receives TF-IDF features, a first hidden layer with 128 neurons (ReLU activation), a second hidden layer with 64 neurons (ReLU activation) and an output layer with 1 neuron (Sigmoid activation) for binary classification

## Number of Epochs
- The model was trained for 15 epochs.

## Activation Functions
- ReLU (Rectified Linear Unit) was used in the hidden layers
- Sigmoid was used in the output layer to produce a probability between 0 and 1

## Model Accuracy
- The model achieved an accuracy of 80% on the test dataset.
